```python
import pandas as pd

data_path = '../dataset/CHD/gear_dataset.csv'
```

```python
df = pd.read_csv(data_path)
```

```python
flagna = df.loc[df['Flag'].isna()]
```

```python
flagna['Data2'].value_counts()
```

```python
flagna['Flag'] = 'R'
```

```python
flagna
```

```python
import numpy as np

flagna.loc[flagna['Data2'] == 'R', 'Data2'] = np.NaN

flagna
```

```python
df = df.dropna(subset='Flag')
```

```python
df = pd.concat([df, flagna])
```

```python
import os

save_path = '../dataset/CHD/modified'

if not os.path.exists(save_path):
    os.makedirs(save_path)
```
